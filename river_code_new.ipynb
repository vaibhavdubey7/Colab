```python
import matplotlib.pyplot as plt
import numpy as np

def estimate_running_cost(vehicle_sales, rider_distance, cost_per_month):
    # Generate 12 monthly samples from the vehicle sales distribution
    monthly_sales = np.random.normal(loc=vehicle_sales.mean(), scale=vehicle_sales.std(), size=12).astype(int)

    # Generate 12 monthly samples from the rider distance distribution
    monthly_distance = np.random.normal(loc=rider_distance.mean(), scale=rider_distance.std(), size=12)

    # Calculate the running cost month-wise
    running_cost = monthly_sales * monthly_distance * cost_per_month

    return running_cost

# Constants
vehicle_sales_mean = 10000
vehicle_sales_std = 0  # Assuming no variability in vehicle sales
rider_distance_mean = 6000
rider_distance_std = 2000
cost_per_month = 3000

# Create normal distributions for vehicle sales and rider distance
vehicle_sales = np.random.normal(loc=vehicle_sales_mean, scale=vehicle_sales_std, size=10000)
rider_distance = np.random.normal(loc=rider_distance_mean, scale=rider_distance_std, size=10000)

# Estimate running cost month-wise
running_cost_monthly = estimate_running_cost(vehicle_sales, rider_distance, cost_per_month)

# Display the estimated running cost month-wise
months = ['April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December', 'January', 'February', 'March']
diwali_month = 'October'  # Update with the Diwali month

for month, cost in zip(months, running_cost_monthly):
    print(f"{month}: INR {cost:.2f}")

# Create a graph of the running cost month-wise
plt.plot(months, running_cost_monthly, marker='o')
plt.xlabel('Month')
plt.ylabel('Running Cost (INR)')
plt.title('Running Cost Month-wise')
plt.grid(True)

# Highlight the Diwali month
diwali_index = months.index(diwali_month)
plt.axvline(x=diwali_index, color='red', linestyle='--', label='Diwali Month')
plt.legend()
plt.xticks(rotation=45, ha='right')

plt.show()
```
